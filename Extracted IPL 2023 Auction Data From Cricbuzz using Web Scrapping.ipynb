```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from IPython.display import FileLink
```

```python
url ="https://m.cricbuzz.com/cricket-series/ipl-2023/auction/completed"
```

```python
html_data = requests.get(url).text
```

```python
html_data
```

```python
html_data[:1000]
```

```python
soup = BeautifulSoup(html_data,'html5lib')
```

```python
player_name_selector = 'h3.cb-list-heading'
player_name_tags =  soup.select(player_name_selector)
```

```python
player_names =[]
for tag in player_name_tags:
    h3_text = ''.join([str(content) for content in tag.contents if content.name != 'span'])
    player_names.append(h3_text.strip())

player_names.pop(0)
```

```python
len(player_names)
```

```python
player_names[:20]
```

```python
player_info_tags= soup.find_all('p',class_="cb-list-intro-text")
```

```python
player_info_list =[]

for tag in player_info_tags:
    player_info_list.append(tag.text)
```

```python
player_info_list
```

```python
player_style = []
player_nationality = []
for data in player_info_list:
    
    split_data = data.split('•')
    split_data = [part.strip() for part in split_data]
    player_style.append(split_data[0])
    player_nationality.append(split_data[1])
    
```

```python
player_style
```

```python
player_nationality
```

```python
auction_tag_list = set()
auction_tag_selector = soup.find_all('div',class_="cb-list-intro-text")

for tags in auction_tag_selector:
    auction_tag_list.add(tags.text)
```

```python
auction_tag_list
```

```python
all_info = []

all_info_tags = soup.find_all('strong')

for tags in all_info_tags:
    all_info.append(tags.text)
```

```python
all_info[0]
```

```python
all_info[1]
```

```python
all_info[2]
```

```python
base_price=[]
final_price=[]
franchise=[]

for index, item in enumerate(all_info):
    if index % 3 == 0:
        base_price.append(item)
    elif index % 3 == 1:
        final_price.append(item)
    elif index % 3 == 2:
        franchise.append(item)    
```

```python
base_price
```

```python
final_price
```

```python
franchise
```

```python
span_tags = soup.find_all('span',class_="pull-right")
```

```python
len(span_tags)
```

```python
player_status = []

for tags in span_tags:
    player_status.append(tags.text)
```

```python
player_status
```

```python
def convert_to_lacs(value_str):
    if value_str == '-':
        return None
    cleaned_value = value_str.replace('\xa0', '').replace('Crore', '').replace('Lakh', '')
    numerical_value = float(cleaned_value.split()[0])
    
    # Check the unit and convert accordingly
    if 'Crore' in value_str:
        return numerical_value * 100  # Convert Crore to Lacs
    elif 'Lakh' in value_str:
        return numerical_value  # Value is already in Lacs
    else:
        return None  # Invalid or unsupported unit

```

```python
base_price_lacs =[]
for value in base_price:
    converted_value = convert_to_lacs(value)
    base_price_lacs.append(converted_value)
print(base_price_lacs)
```

```python
final_price_lacs =[]
for value in final_price:
    converted_value = convert_to_lacs(value)
    final_price_lacs.append(converted_value)
print(final_price_lacs)
```

```python
df = pd.DataFrame({
    'player name':player_names,
    'player role':player_style,
    'player nationality':player_nationality,
    'base price(in lakhs)':base_price_lacs,
    'final price(in lakhs)':final_price_lacs,
    'player status':player_status,
    'franchise name':franchise

})
```
